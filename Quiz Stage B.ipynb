```python
#Q13
from sklearn.preprocessing import MinMaxScaler
df1=df1.drop(columns=['date','lights'])
scaler= MinMaxScaler()
normalised_df=pd.DataFrame(scaler.fit_transform(df1),columns=df1.columns)

features_df=normalised_df.drop(columns=['Appliances'])
heating_target=normalised_df['Appliances']

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(features_df,heating_target,test_size=0.3, random_state=42)

from sklearn import datasets, linear_model, metrics
reg=linear_model.LinearRegression()
reg.fit(x_train,y_train)
predicted_values= reg.predict(x_test)


#MAE
from sklearn.metrics import mean_absolute_error
mae= mean_absolute_error(y_test,predicted_values)
round(mae, 2)
```

```python
#Q17
from sklearn.linear_model import LinearRegression 
def get_weights_df(model,feat,col_name):
    weights=pd.Series(model.coef_,feat.columns).sort_values()
    weights_df= pd.DataFrame(weights).reset_index()
    weights_df.columns=['Features',col_name]
    weights_df[col_name].round(3)
    return weights_df
reg_weights= get_weights_df(reg,x_train,'Linear_Model_Weight')
reg_weights
```

```python
#Q18
from sklearn.linear_model import Ridge
ridge_reg= Ridge(alpha=0.4)
ridge_reg.fit(x_train,y_train)

ridge_predicted_values= ridge_reg.predict(x_test)

rmse=np.sqrt(mean_squared_error(y_test,ridge_predicted_values))
round(rmse,3)
```
